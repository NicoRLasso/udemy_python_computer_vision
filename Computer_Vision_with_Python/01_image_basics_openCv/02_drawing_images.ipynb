```python
import numpy as np
import cv2

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
blank_image =np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_image.shape
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(384,10),pt2=(505,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image,center=(400,400),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blank_image)
```

```python
cv2.line(img=blank_image,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_image)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_image,"Hello",org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_image)
```

```python
blank_image = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_image)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(blank_image,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_image)
```
