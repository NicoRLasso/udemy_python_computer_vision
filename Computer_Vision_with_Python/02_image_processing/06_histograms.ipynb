```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
dark_horse = cv2.imread("../DATA/horse.jpg") # Original BGR OpenCV
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # Converted RGB OpenCV

rainbow = cv2.imread("../DATA/rainbow.jpg") # Original BGR OpenCV
show_rainbow = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # Converted RGB OpenCV

blue_bricks = cv2.imread("../DATA/bricks.jpg") # Original BGR OpenCV
show_bricks = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # Converted RGB OpenCV
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
# OpenCV BGR
hist_values = cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
# OpenCV BGR
hist_values = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
img = blue_bricks
color = ("b","g","r")

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,col)
    plt.xlim([0,256])

plt.title("HISTOGRAM FOR BLUE BRICK")
```

```python
img = dark_horse
color = ("b","g","r")

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,col)
    plt.xlim([0,256])
    plt.ylim([0,500000])

plt.title("HISTOGRAM FOR DARK HORSE")
```
